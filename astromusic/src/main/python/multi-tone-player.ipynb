```python
def amp(a):
    # a from [0, 1]: 
    # 1 -> MAX_AMPL = 2**15 - 1
    # 0.9 -> MAX_AMPL / 2
    # ...
    # 0.1 -> MAX_AMPL / 2**9
    # 0.0 -> 0
    MAX_AMPL = 2**15 - 1 
    ascaled = MAX_AMPL * 0.5**(10*(1-a))
    ascaled[a==0] = 0
    return ascaled 

aa = np.linspace(0,1,11)
plt.figure()
plt.plot(aa, amp(aa))
```
